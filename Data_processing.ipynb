```python
import pandas as pd
import numpy as np
```

```python
ranking_df = pd.read_csv("data.csv")
ranking_df.head()
```

```python
ranking_df = ranking_df[ranking_df['Position'] < 101]
ranking_df.shape
```

```python
ranking_df['Date'] = pd.to_datetime(ranking_df["Date"])
```

```python
ranking_df['Week'] = ranking_df['Date'].dt.to_period('W')
```

```python
ranking_weekly_df = ranking_df.groupby(['Region', 'Week', 'Track Name', 'Artist', 'URL']).sum()
```

```python
ranking_weekly = ranking_weekly_df.reset_index()
ranking_weekly.sort_values(['Region', 'Week', 'Streams'], ascending=False, inplace=True)
ranking_weekly.set_index(['Region', 'Week'], inplace=True)
ranking_weekly.drop('Position', axis=1, inplace=True)
ranking_weekly.head()
```

```python
ranking_weekly_new_position = pd.DataFrame()
for region, region_df in ranking_weekly.groupby(level=0):
    for week, week_df in region_df.groupby(level=1):
        print(region, week)
        week_df['Position'] = np.arange(1, week_df.shape[0]+1)
        ranking_weekly_new_position = pd.concat([ranking_weekly_new_position, week_df])
        
        
```

```python
ranking_weekly_new_position
```

```python
ranking_weekly_new_position_top100 = ranking_weekly_new_position[ranking_weekly_new_position['Position'] < 101]
```

```python
ranking_weekly_new_position_top100.to_csv("ranking_weekly_top100.csv")
```
